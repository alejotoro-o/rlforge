```python
import numpy as np
import matplotlib.pyplot as plt

import gymnasium as gym

from rlforge.experiments import runEpisodicExperiment
from rlforge.agents.semi_gradient import DQNAgent
```

```python
env = gym.make("MountainCar-v0")
agent = DQNAgent(learning_rate=0.001,
                 discount=0.99,
                 state_dim=2,
                 num_actions=env.action_space.n,
                 temperature=0.001,
                 network_architecture=[256],
                 target_network_update_steps=1,
                 num_replay=4,
                 experience_buffer_size=50000,
                 mini_batch_size=8)
```

```python
rewards, steps_per_episode = runEpisodicExperiment(env, agent, 10, 50, 10000)
```

```python
plt.plot(np.mean(rewards, axis=1))
plt.grid()
```

```python
plt.plot(np.mean(steps_per_episode, axis=1))
plt.grid()
```
